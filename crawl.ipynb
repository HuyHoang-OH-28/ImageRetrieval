```python
from bs4 import BeautifulSoup  #parse HTML
from urllib.parse import urljoin, urlparse #Check link co hay khong?
import urllib.request, urllib.error #Tao request tai anh ve folder
import threading  # lap trinh multi threading
import time #sleep, tinh toan time
import random
import os
import requests
```

# Get All urls from link

```python
def is_valid(url):
    parsed = urlparse(url)
    return bool(parsed.netloc) and bool(parsed.scheme)
```

```python
def get_all_images(url,slug):
    soup = BeautifulSoup(urllib.request.urlopen(url),"html.parser")
    urls = []
    for img in soup.find_all("img"):
        img_url = img.attrs.get("src")
        img_alt = img.attrs.get("alt")
        #print(img_alt)
        #print(img_url)
        try:
            pos = img_alt.index(slug)
            if pos >=0: 
                if not img_url:
                    continue
                img_url = urljoin(url,img_url)
                try:
                    pos = img_url.index("?")
                    img_url = img_url[:pos]
                except ValueError:
                    pass

                if is_valid(img_url):
                    urls.append(img_url)
        except ValueError:
            pass
    return urls
```

# Create file txt to save all url of image orderby Category


```python
class checkTimeThread():
    def __init__(self,n_Threads,n_pages,url_topic,topic):
        self.n_Threads = n_Threads
        self.ExcuteThread = None
        self.result = []
        self.n_pages = n_pages
        self.url_topic = url_topic
        self.topic = topic
        
    def main(self,start,end): 
        for i in range(start,end):
            try:
                self.result.extend(get_all_images(self.url_topic+str(i),self.topic))
            except:
                pass
            
    def __call__(self):
        
        threads = []
    
        batch_size = self.n_pages // self.n_Threads # so page trong 1 thread 
        for i in range(0,self.n_pages,batch_size):
            start = i
            end = i + batch_size
            
            if end >= self.n_pages:
                end = self.n_pages +1
            
            threads.append(threading.Thread(target=self.main,args = (start,end)))
            
        start = time.time()
        for i in range(self.n_Threads):
            threads[i].start()
        for i in range(self.n_Threads):
            threads[i].join()
        end = time.time()
        
        self.ExcuteThread = end -start
        
        return self.result
    def __str__(self):
        return f"Excute with {self.n_Threads} Threads = {self.ExcuteThread} s\n"
              
```

```python
n_Threads = 30
n_pages = 8
for key in (dic_categories):
    for name in dic_categories[key]:
        link = f"https://www.freepik.com/search?from_query={name}&query={name}&sort=popular&type=photo&page="
        result = checkTimeThread(min(n_Threads, n_pages//2),n_pages,link,name.lower())() 
        result2 = list(set(result))
        print(f"\n{key}_{name}.txt have {len(result2)} images")

        strResult = '\n'.join(result2)
        with open(f"TXT_Files/{key}_{name}.txt", "w") as f:
            f.write(strResult)
```

# Download image from txt file


```python
class DownloadFromUrl():
    def __init__(self, n_Threads,urls,destinate_folder):
        self.n_Threads = n_Threads
        self.urls = urls
        self.n = len(urls)
        self.destinate_folder = destinate_folder
        if not os.path.isdir(destinate_folder):
            os.mkdir(destinate_folder)
            
        
    def down_url(self, start, end):

        for i in range(start, end):
            try:
                response=requests.get(self.urls[i])
            except Exception as e:
                continue
           
            # Tránh trùng tên
            a = random.random()
            with open(f"{self.destinate_folder}/{a}.jpg", "wb") as f:
                f.write(response.content)      
            
    def __call__(self):
        try:
            
            threads = []
            
            #=================
            batch = self.n // self.n_Threads
            
            for i in range(0, self.n, batch):
                start = i
                end = i + batch

                if end >= self.n:
                    end = self.n + 1

                threads.append(threading.Thread(target=self.down_url, args = (start, end)))
        
            #=================
            for i in range(self.n_Threads):
                threads[i].start()
            for i in range(self.n_Threads):
                threads[i].join()

        except Exception as e:
            pass    
       
    def __str__(self):
        return f"Excute with {self.n_Threads} Thread = {self.ExcuteThread} s\n"
        
```

```python
l = []
for name_txt in os.listdir("TXT_Files"):
    if name_txt.endswith(".txt"):
        l.append(name_txt)
```
