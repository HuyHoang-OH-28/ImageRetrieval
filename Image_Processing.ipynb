# Preprocessing data


```python
root_img_path = "DataBase"
dic_categories = ['animal','scenery', 'furniture', 'plant']
count = 0
```

```python
for folder in os.listdir(root_img_path):
    print(folder)
    if folder.split('_')[0] in dic_categories:
        path = root_img_path +"/"+ folder
        list_imgs = [path+'/'+name for name in os.listdir(path) if name.endswith((".jpg", ".png", ".jpeg"))]
        count_img = 0
        for path_img in list_imgs:
            try:
                img = Image.open(path_img)
                if img.width < 10:
                    print("Image too small: ", path_img)
                    os.remove(path_img) # chú ý chỗ này, mới đầu chạy phải comment lại không lỡ lỗi mất hết ảnh
                    continue
                img.verify() # verify that it is, in fact an image
                
                if count_img > 200:
                    break
                #print(path_img)
                img = Image.open(path_img).resize((224, 224))
        
                
             
                # lỗi không xác đinh
                
                
                count_img +=1
                if not os.path.isdir("Img_Processing/"+ folder):
                    os.mkdir("Img_Processing/"+ folder)
                img = img.save("Img_Processing/"+ folder+"/"+folder.split('_')[1]+str(count_img)+".jpg")
            except Exception:
                count += 1
                print("error: ", path_img) 
```
